## Clean Data

```python
# Several columns have a high percentage of nulls; drop those with > 40% nulls
null_counts = cc.isnull().sum()
null_df = null_counts.sort_values(ascending=False).to_frame(name="null_count")
null_df["null_pct"] = (null_df["null_count"] / len(cc) * 100).round(2)

for col in null_df.index:
    if null_df.loc[col, "null_pct"] > 40:
        print(
            f"Dropping column '{col}' with {null_df.loc[col, 'null_pct']}% nulls"
        )
        cc.drop(columns=[col], inplace=True)

cc.head()
```

```python
# Several columns have a high percentage of zeros; drop those with > 20% zeros
zero_counts = (cc == 0).sum()
zero_percentage = zero_counts / len(cc)
columns_to_drop = zero_percentage[zero_percentage > 0.2].index
print(f"Dropping columns: {list(columns_to_drop)}")
cc.drop(columns=columns_to_drop, inplace=True)
cc.head()
```

```python
# Normalize column names to `snake_case`
cc.columns = [
    col.lower().replace(".", "_").replace("-", "_").replace(" ", "_")
    for col in cc.columns
]
cc.head()
```

```python
# For each college, interpolate missing numeric values over over the years, but
# do not fill in trailing nulls (i.e., if the last few years are missing, keep
# them missing)

num_cols = cc.select_dtypes(include='number').columns.tolist()
cc = cc.sort_values(["college", "year"])

def interpolate_backward_allowed(g: pd.DataFrame) -> pd.DataFrame:
    g_orig = g.copy()

    # 1. Forward interpolation (normal behavior)
    fwd = g.interpolate(method="linear", limit_direction="forward")

    # 2. Backward interpolation — fills earlier values but should NOT fill trailing values
    bwd = fwd.interpolate(method="linear", limit_direction="backward")

    # 3. Identify trailing null regions (per column)
    for col in g.columns:
        # Last valid index in original data
        last_valid = g_orig[col].last_valid_index()
        if last_valid is not None:
            # Restore everything AFTER the last valid value to null
            bwd.loc[last_valid+1:, col] = g_orig.loc[last_valid+1:, col]

    return bwd

cc[num_cols] = (
    cc.groupby("college")[num_cols]
      .apply(interpolate_backward_allowed)
      .reset_index(drop=True)
)
```

```python
# Now, if any column is null for a particular college in a particular year,
# replace it with the average of the non-null values in that column for all
# colleges over that year.

# For each year, compute the column-wise means across *all colleges*
yearly_means = cc.groupby("year")[num_cols].transform("mean")

# Replace remaining nulls
cc[num_cols] = cc[num_cols].fillna(yearly_means)
```

## Regularized linear regression (Elastic Net)

If $y_i = \sum_{j=1}^p \beta_{j}x_{ij}$, the the loss function takes the form:
$$
  L = \frac{1}{2n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
    + \alpha \left(
        \ell_1 \sum_{i=1}^n |\beta_j|
        + (1 - \ell_1) \sum_{j=1}^n \beta_j^2
    \right)
$$
where $\alpha$ is the regularization parameter and $\ell_1$ (called `l1_ratio` in `sklearn`) is the mixing parameter.

Pros:
* Handles multicollinearity well.
* Easy to interpret.

```python
pipeline = Pipeline(
    [
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler()),
        ("elasticnet", ElasticNet(max_iter=10000, random_state=2025)),
    ]
)

param_grid = {
    "elasticnet__alpha": np.logspace(-4, 4, 100),
    "elasticnet__l1_ratio": np.linspace(0.001, 0.5, 100),
}
```

```python
train_sizes = np.linspace(
    int(0.1 * len(X_train)),
    len(X_train),
    10,
    dtype=int,
)

cv_rmses = []

for n in train_sizes:
    X_sub, y_sub = resample(
        X_train,
        y_train,
        n_samples=n,
        random_state=2025,
        replace=False,
    )

    grid = GridSearchCV(
        pipeline,
        param_grid,
        cv=5,
        scoring="neg_mean_squared_error",
        n_jobs=-1,
    )

    grid.fit(X_sub, y_sub)

    # Best CV MSE is negative, so negate and sqrt
    best_cv_rmse = np.sqrt(-grid.best_score_)
    cv_rmses.append(best_cv_rmse)

    print(f"N={n:<6}  CV RMSE={best_cv_rmse:.4f}")
```

```python
def cv_learning_curve(
    X: np.ndarray,
    y: np.ndarray,
    train_sizes: np.ndarray,
    n_repeats: int = 5,
    cv: int = 5,
    random_state: int = 2025,
):
    """
    Compute a cross-validated learning curve for a given model pipeline.

    :param X: Feature matrix
    :param y: Target vector
    :param train_sizes: List or array of training set sizes to evaluate
    :param n_repeats: Number of repetitions for each training size
    :param cv: Number of cross-validation folds
    :param random_state: Random seed for reproducibility
    """
    
    rng = np.random.RandomState(random_state)
    rmses = np.zeros((len(train_sizes), n_repeats))

    for i, n in enumerate(train_sizes):
        for r in range(n_repeats):
            X_sub, y_sub = resample(
                X,
                y,
                n_samples=n,
                replace=False,
                random_state=rng.randint(1_000_000),
            )

            grid = GridSearchCV(
                pipeline,
                param_grid,
                cv=cv,
                scoring="neg_mean_squared_error",
                n_jobs=-1,
            )

            grid.fit(X_sub, y_sub)
            rmses[i, r] = np.sqrt(-grid.best_score_)

        print(
            f"N={n:<6}  "
            f"RMSE={rmses[i].mean():.4f} ± {rmses[i].std(ddof=1)/np.sqrt(n_repeats):.4f}"
        )
    
    return rmses


train_sizes = np.linspace(
    int(0.1 * len(X_train)),
    len(X_train),
    10,
    dtype=int,
)

rmse_matrix = cv_learning_curve(
    X_train,
    y_train,
    train_sizes=train_sizes,
    n_repeats=5,
)
```

```python
mean_rmse = rmse_matrix.mean(axis=1)
se_rmse = rmse_matrix.std(axis=1, ddof=1) / np.sqrt(rmse_matrix.shape[1])

plt.figure(figsize=(7.5, 5))

plt.plot(
    train_sizes,
    mean_rmse,
    linewidth=2.2,
    label="Cross-validated RMSE",
)

plt.fill_between(
    train_sizes,
    mean_rmse - se_rmse,
    mean_rmse + se_rmse,
    alpha=0.25,
    label="±1 SE",
)

plt.xlabel("Training Sample Size", fontsize=12)
plt.ylabel("RMSE", fontsize=12)
plt.title(
    "Elastic Net Learning Curve (CV RMSE)",
    fontsize=13,
)

plt.legend(frameon=False)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from matplotlib.patches import Patch


def extract_elasticnet_importance(
    fitted_pipeline,
    feature_names,
):
    """
    Returns a DataFrame with standardized coefficients
    from a fitted Elastic Net pipeline.
    """
    coefs = fitted_pipeline.named_steps["elasticnet"].coef_

    importance = (
        pd.DataFrame(
            {
                "feature": feature_names,
                "coef": coefs,
                "abs_coef": np.abs(coefs),
            }
        )
        .query("abs_coef > 0")   # drop zeroed features
        .sort_values("abs_coef", ascending=False)
        .reset_index(drop=True)
    )

    return importance


importance_df = extract_elasticnet_importance(
    fitted_pipeline=grid.best_estimator_,
    feature_names=X_train.columns,
)

top_k = 15
importance_top = importance_df.head(top_k)



colors = np.where(
    importance_top["coef"] > 0,
    "tab:blue",
    "tab:orange",
)

plt.figure(figsize=(7.5, 0.35 * len(importance_top)))

plt.barh(
    importance_top["feature"],
    importance_top["abs_coef"],
    color=colors,
)

plt.gca().invert_yaxis()
plt.xlabel("Absolute Standardized Coefficient", fontsize=12)
plt.title("Elastic Net Feature Importance", fontsize=13)

legend_elements = [
    Patch(facecolor="tab:blue", label="Positive effect"),
    Patch(facecolor="tab:orange", label="Negative effect"),
]
plt.legend(handles=legend_elements, frameon=False)

plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import RandomizedSearchCV
import joblib
import numpy as np
import matplotlib.pyplot as plt
```

```python
rf_pipeline = Pipeline(
    [
        ("imputer", SimpleImputer(strategy="mean")),
        ("rf", RandomForestRegressor(random_state=42, n_jobs=-1)),
    ]
)

param_dist = {
    "rf__n_estimators": [100, 200, 300, 500],
    "rf__max_depth": [None, 10, 20, 30, 50],
    "rf__min_samples_split": [2, 5, 10],
    "rf__min_samples_leaf": [1, 2, 4],
    "rf__max_features": ["sqrt", "log2", 0.2, 0.5],
}

rf_search = RandomizedSearchCV(
    rf_pipeline,
    param_distributions=param_dist,
    n_iter=30,
    cv=5,
    scoring="neg_mean_squared_error",
    n_jobs=-1,
    random_state=42,
    verbose=1,
)

rf_search.fit(X_train, y_train)

best_pipeline = rf_search.best_estimator_
print("Best params:", rf_search.best_params_)


def eval_split(model, X, y, name):
    preds = model.predict(X)
    mse = mean_squared_error(y, preds)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, preds)
    print(f"{name:10s} -> RMSE: {rmse:.6f}, R2: {r2:.4f}")
    return preds

_ = eval_split(best_pipeline, X_train, y_train, "Train")
if not X_valid.isnull().all().all():
    _ = eval_split(best_pipeline, X_valid, y_valid, "Valid")
_ = eval_split(best_pipeline, X_test, y_test, "Test")

# Feature importances (from the fitted RandomForest inside the pipeline)
rf_model = best_pipeline.named_steps["rf"]
fi = pd.Series(
    rf_model.feature_importances_, index=X_train.columns
).sort_values(ascending=False)

print("Top 20 feature importances:")
fi.head(20)

# # Save fitted pipeline
# joblib.dump(best_pipeline, "random_forest_pipeline.joblib")
# print("Saved random_forest_pipeline.joblib")
```

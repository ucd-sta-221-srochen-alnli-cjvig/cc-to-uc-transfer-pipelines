## Clean Data

```python
# Several columns have a high percentage of nulls; drop those with > 40% nulls
null_counts = cc.isnull().sum()
null_df = null_counts.sort_values(ascending=False).to_frame(name="null_count")
null_df["null_pct"] = (null_df["null_count"] / len(cc) * 100).round(2)

for col in null_df.index:
    if null_df.loc[col, "null_pct"] > 40:
        print(
            f"Dropping column '{col}' with {null_df.loc[col, 'null_pct']}% nulls"
        )
        cc.drop(columns=[col], inplace=True)

cc.head()
```

```python
# Several columns have a high percentage of zeros; drop those with > 20% zeros
zero_counts = (cc == 0).sum()
zero_percentage = zero_counts / len(cc)
columns_to_drop = zero_percentage[zero_percentage > 0.2].index
print(f"Dropping columns: {list(columns_to_drop)}")
cc.drop(columns=columns_to_drop, inplace=True)
cc.head()
```

```python
# Normalize column names to `snake_case`
cc.columns = [
    col.lower().replace(".", "_").replace("-", "_").replace(" ", "_")
    for col in cc.columns
]
cc.head()
```

```python
# For each college, interpolate missing numeric values over over the years, but
# do not fill in trailing nulls (i.e., if the last few years are missing, keep
# them missing)

num_cols = cc.select_dtypes(include='number').columns.tolist()
cc = cc.sort_values(["college", "year"])

def interpolate_backward_allowed(g: pd.DataFrame) -> pd.DataFrame:
    g_orig = g.copy()

    # 1. Forward interpolation (normal behavior)
    fwd = g.interpolate(method="linear", limit_direction="forward")

    # 2. Backward interpolation — fills earlier values but should NOT fill trailing values
    bwd = fwd.interpolate(method="linear", limit_direction="backward")

    # 3. Identify trailing null regions (per column)
    for col in g.columns:
        # Last valid index in original data
        last_valid = g_orig[col].last_valid_index()
        if last_valid is not None:
            # Restore everything AFTER the last valid value to null
            bwd.loc[last_valid+1:, col] = g_orig.loc[last_valid+1:, col]

    return bwd

cc[num_cols] = (
    cc.groupby("college")[num_cols]
      .apply(interpolate_backward_allowed)
      .reset_index(drop=True)
)
```

```python
# Now, if any column is null for a particular college in a particular year,
# replace it with the average of the non-null values in that column for all
# colleges over that year.

# For each year, compute the column-wise means across *all colleges*
yearly_means = cc.groupby("year")[num_cols].transform("mean")

# Replace remaining nulls
cc[num_cols] = cc[num_cols].fillna(yearly_means)
```

```python
print("Old shape:", cc.shape)

# Remove outliers in the target variable based on 2 standard deviations in any
# given reporting year

col = "completion_title_iv_transf_completed_4yr_by_4yrs"
stats = cc.groupby("year")[col].agg(mean="mean", std="std").reset_index()
stats["std"] = stats["std"].fillna(0)

cc = cc.merge(stats, on="year", how="left")
mask = (cc["std"] == 0) | (
    (cc[col] >= cc["mean"] - 2 * cc["std"])
    & (cc[col] <= cc["mean"] + 2 * cc["std"])
)
cc = cc[mask].drop(columns=["mean", "std"])

print("New shape:", cc.shape)
```

```python
# Since four-year completion rates cannot be known until four years after
# transfer, I cannot use the same years for training, validation, and testing
# as everyone else did for Q1 and Q3.

# train_years = list(range(2012, 2020))
# valid_years = [2020, 2021]
# test_years = [2022, 2023]

train_years = list(range(2012, 2018))
valid_years = [2018]
test_years = [2019]

train = cc[cc["year"].isin(train_years)]
valid = cc[cc["year"].isin(valid_years)]
test  = cc[cc["year"].isin(test_years)]

excluded_cols = [
    "college",
    "year",
    "completion_title_iv_transf_completed_4yr_by_4yrs"
]

feature_cols = [c for c in cc.columns if c not in excluded_cols]

X_train = train[feature_cols]
y_train = train["completion_title_iv_transf_completed_4yr_by_4yrs"]

X_valid = valid[feature_cols]
y_valid = valid["completion_title_iv_transf_completed_4yr_by_4yrs"]

X_test = test[feature_cols]
y_test = test["completion_title_iv_transf_completed_4yr_by_4yrs"]
```

# Training

```python
import joblib
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import (
    mean_squared_error, r2_score
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

## Regularized linear regression (Elastic Net)

If $y_i = \sum_{j=1}^p \beta_{j}x_{ij}$, the the loss function takes the form:
$$
  L = \frac{1}{2n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
    + \alpha \left(
        \ell_1 \sum_{i=1}^n |\beta_j|
        + (1 - \ell_1) \sum_{j=1}^n \beta_j^2
    \right)
$$
where $\alpha$ is the regularization parameter and $\ell_1$ (called `l1_ratio` in `sklearn`) is the mixing parameter.

Pros:
* Handles multicollinearity well.
* Easy to interpret.

```python
# Build a pipeline: 
# impute -> scale -> elastic net with CV over l1_ratio and alpha
l1_ratios = np.linspace(0.1, 0.9, 9)
enet_cv = ElasticNetCV(
    l1_ratio=l1_ratios,
    n_alphas=100,
    cv=5,
    max_iter=10000,
    n_jobs=-1,
    random_state=42,
    tol=1e-4,
)

pipeline = Pipeline(
    [
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler()),
        ("elasticnet", enet_cv),
    ]
)

pipeline.fit(X_train, y_train)

# Extract fitted estimator and best hyperparameters
fitted_enet: ElasticNetCV = pipeline.named_steps["elasticnet"]
best_alpha = fitted_enet.alpha_
best_l1_ratio = fitted_enet.l1_ratio_
print(f"Best alpha: {best_alpha:.6g}, Best l1_ratio: {best_l1_ratio:.3f}")

# Predictions and evaluation
def evaluate(X, y, split_name):
    preds = pipeline.predict(X)
    mse = mean_squared_error(y, preds)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, preds)
    print(f"{split_name}".ljust(10), f"-> RMSE: {rmse:.6f}, R2: {r2:.4f}")
    return preds

_ = evaluate(X_train, y_train, "Train")
if not X_valid.isnull().all().all():
    _ = evaluate(X_valid, y_valid, "Validation")
_ = evaluate(X_test, y_test, "Test")

# joblib.dump(pipeline, "elasticnet_pipeline.joblib")
# print("Saved pipeline to elasticnet_pipeline.joblib")
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Plot learning curves
train_sizes, train_scores, test_scores = learning_curve(
    pipeline,
    X_train,
    y_train,
    cv=5,
    scoring="neg_mean_squared_error",
    n_jobs=-1
)

train_rmse = np.sqrt(-train_scores.mean(axis=1))
test_rmse = np.sqrt(-test_scores.mean(axis=1))

plt.plot(train_sizes, train_rmse, label="Train RMSE", color="blue")
plt.plot(train_sizes, test_rmse, label="Test RMSE", color="red")
plt.xlabel("Training Size")
plt.ylabel("RMSE")
plt.title("Learning Curve")
plt.legend()
plt.show()
```
